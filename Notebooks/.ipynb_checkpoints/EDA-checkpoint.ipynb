1. First we need to convert that reference date to datetime, because you want to perform date operations on it.
Think about it: What function in pandas is there that converts the string ``2025-07-21'' to a datetime object?
2. Now for each row (customer), you want to subtract the number of months from that date.
Question: Does pandas have a function that can subtract "a certain number of months" from a date? Not days, not weeks, just months?
3.Now that you have found the right function to subtract the month, how do you apply it to all customers (all rows in the dataframe)?
For example, with `.apply()` or with a direct vector operation.

````python
import pandas as pd

# Assuming df is already defined with a 'tenure' column containing numeric values
current_date = pd.to_datetime("2025-07-21")

# Make sure tenure column contains numeric values and handle any non-numeric values
df['tenure'] = pd.to_numeric(df['tenure'], errors='coerce')

# Convert tenure to integer before applying DateOffset
# Using fillna(0) to handle any NaN values that might result from the conversion
df['start_date'] = df['tenure'].fillna(0).apply(lambda x: current_date - pd.DateOffset(months=int(x)))
````


### ✅ نسخه اول:

```python
df['end_date'] = df['Churn'].apply(
    lambda x: current_date if x == 'No' else current_date - pd.DateOffset(months=1)
)
```

یعنی:
- اگر مشتری هنوز فعاله (`Churn = 'No'`) → تاریخ پایان = امروز
- اگر مشتری ترک کرده (`Churn = 'Yes'`) → فرض می‌کنیم **یک ماه پیش** رفته

📌 این برای تحلیل‌هایی مثل "مشتری‌های تازه رفته" یا "سه ماهه خرید نکرده" مناسب‌تره چون تلاش می‌کنی زمان خروج واقعی رو تخمین بزنی.

---

### ✅ نسخه دوم:

```python
df['end_date'] = df['Churn'].apply(
    lambda x: current_date if x == 'No' else current_date + pd.DateOffset(years=1)
)
```

یعنی:
- مشتری فعاله → پایان = امروز
- مشتری ترک کرده → پایان = یک سال بعد از امروز 😬

📌 این نسخه یه جورایی **قراردادی** حساب کرده: انگار قراردادی که ترک شده بود، یک‌ساله بوده و هنوز باید تا سال دیگه ادامه پیدا می‌کرد—اما این با `Churn = 'Yes'` تناقض داره، چون اون مشتری الان دیگه عضو نیست.

باشه حامد عزیز، الان همون کدی که با پانداز (pandas) نوشتم رو مرحله‌به‌مرحله توضیح می‌دم تا کامل متوجه بشی چی داره اتفاق می‌افته. این همون کد بود:

```python
result = df.groupby('Contract').agg(
    total_customers=('Contract', 'count'),
    churned_customers=('Churn', lambda x: (x == 'Yes').sum())
)

result['churn_rate_percent'] = (result['churned_customers'] / result['total_customers'] * 100).round(2)
```

حالا بریم تو دلش مرحله‌به‌مرحله ⬇️

---

### 🧩 مرحله ۱: گروه‌بندی داده‌ها بر اساس نوع قرارداد

```python
df.groupby('Contract')
```

یعنی کل دیتای مشتری‌ها رو تقسیم می‌کنیم به چند دسته، بر اساس ستون `Contract`:
- `'Month-to-month'`
- `'One year'`
- `'Two year'`

هر دسته مخصوص یک نوع قرارداد می‌شه.

---

### 🧮 مرحله ۲: محاسبه تعداد کل مشتری‌ها و تعداد ترک‌کننده‌ها برای هر قرارداد

```python
.agg(
    total_customers=('Contract', 'count'),
    churned_customers=('Churn', lambda x: (x == 'Yes').sum())
)
```

دو تا مقدار برای هر گروه حساب می‌کنیم:

- **total_customers**: با `'count'` می‌شماریم چند نفر تو اون نوع قرارداد بودن
- **churned_customers**: با `lambda x: (x == 'Yes').sum()` می‌شماریم چند نفر تو اون گروه قرارداد، مقدار `Churn`شون `'Yes'` بوده (یعنی سرویس رو قطع کردن)

---

### 🎯 مرحله ۳: محاسبه درصد ترک مشتری‌ها

```python
result['churn_rate_percent'] = (result['churned_customers'] / result['total_customers'] * 100).round(2)
```

اینجا یه ستون جدید توی DataFrame اضافه می‌کنیم:

- نرخ ترک = تعداد ترک‌کننده‌ها تقسیم بر کل مشتری‌ها، ضربدر ۱۰۰
- بعد با `.round(2)` مقدار نهایی رو تا دو رقم اعشار گرد می‌کنیم

---

### 📊 مرحله ۴: نمایش خروجی

```python
result
```

اگه بخوای بگم چطوری همین کد رو به نمودار تبدیل کنی، یا توش شرط‌ها و ستون‌های بیشتری بیاری، من پایه‌ام حامد جان. فقط بگو کجا ببریمش بعدش 😎📈


````python
result = df.groupby('Contract').agg(
    total_customers=('Contract', 'count'),
    churned_customers=('Churn', lambda x: (x == 1).sum())
)

result['Churn_rate_percent'] = (result['churned_customers'] / result['total_customers'] * 100).round(2)
````
